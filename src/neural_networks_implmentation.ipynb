```python
label_name = []
features = []

for class_name in os.listdir(os.getcwd() + r"\data\cat_dog_tiger"):
    #print(class_name)
    for extract_feature in os.listdir(os.getcwd() + r"\data\cat_dog_tiger\\"+class_name):
        features.append(os.getcwd() + "\data\cat_dog_tiger\\"+ class_name + "\\" + extract_feature)
        label_name.append(class_name)
```

```python
features
```

```python
label_name
```

```python
import cv2 
import matplotlib.pyplot as plt 

sample_image = features[2]
sample_img = cv2.imread(sample_image)
plt.imshow(sample_img)
plt.xlabel(label_name[2])
plt.show()
```

```python
import numpy as np

preprocess_image = []
for read_image in features:
    form_image = cv2.imread(read_image)
    resize_image = cv2.resize(form_image, (32,32))
    array_image = np.array(resize_image)
    print(array_image.shape)
    preprocess_image.append(array_image)
```

```python
preprocess_image[0]
```

```python
from sklearn.preprocessing import LabelEncoder 
from tensorflow.keras.utils import to_categorical

encoder = LabelEncoder()
labels_int = encoder.fit_transform(label_name)
labels_onehot = to_categorical(labels_int)
labels_onehot
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(preprocess_image, labels_onehot, test_size=0.3,random_state=0)

x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
import tensorflow as tf 

model_structure = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64,activation = tf.nn.relu),
    tf.keras.layers.Dense(32,activation = tf.nn.relu),
    tf.keras.layers.Dense(16,activation = tf.nn.relu),
    tf.keras.layers.Dense(3,activation = tf.nn.softmax) # output
])
```

```python
model_structure.summary()
```

```python
model_structure.compile(optimizer="adam", loss="categorical_crossentropy",  metrics =["acc"])
```

```python
batch_size = 20

history = model_structure.fit(x=x_train,y=y_train, validation_data=[x_test,y_test],  epochs=10)
```

```python
evl_loss, eval_acc = model_structure.evaluate(x_test,y_test)
print(evl_loss)
print(eval_acc)
```

```python
test_image = cv2.imread(os.getcwd() + '\\data' + "\\images.jpg")
test_image = cv2.resize(test_image, (32,32))
array_test_image = np.array([test_image])
print(array_test_image.shape)
```

```python
categories = ['cat', 'dog', 'tiger']
result = model_structure.predict(array_test_image)
result = np.argmax([result])
predicted_label = categories[result]
predicted_label
```
