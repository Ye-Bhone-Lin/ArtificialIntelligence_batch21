```python
import pandas as pd
import numpy as np
```

```python
unicode_list = []

with open("/content/unicode.unicode", "r", encoding='utf8') as unicode_file:
  for i in unicode_file:
    unicode_list.append(i)
```

```python
unicode_list
```

```python
zawgyi_list = []

with open("/content/zawgyi.zawgyi", "r", encoding='utf8') as zawgyi_file:
  for i in zawgyi_file:
    zawgyi_list.append(i)

zawgyi_list
```

```python
unicode_df= pd.DataFrame()
zawgyi_df = pd.DataFrame()
```

```python
unicode_df['message'] = unicode_list
zawgyi_df['message'] = zawgyi_list
```

```python
zawgyi_df[0:3]
```

```python
unicode_df.shape[0]
```

```python
unicode_df["label"] = [1] * unicode_df.shape[0]
unicode_df
```

```python
zawgyi_df["label"] = [0] * zawgyi_df.shape[0]
zawgyi_df
```

```python
train_unicode_size = int(unicode_df.shape[0]*0.7)
train_unicode_size
```

```python
train_unicode = unicode_df.head(train_unicode_size)
train_unicode
```

```python
test_unicode = unicode_df.tail(unicode_df.shape[0] - train_unicode_size)
test_unicode
```

```python
train_zawgyi_size = int(zawgyi_df.shape[0]*0.7)
train_zawgyi = zawgyi_df.head(train_zawgyi_size)
test_zawgyi = zawgyi_df.tail(zawgyi_df.shape[0] - train_zawgyi_size)
test_zawgyi
```

```python
train = pd.concat([train_unicode, train_zawgyi], axis=0)
test = pd.concat([test_unicode, test_zawgyi], axis=0)
```

```python
train = train.sample(frac=1)
test = test.sample(frac=1)
```

```python
train
```

```python
import re

train['message'] = [i.strip().lower() for i in train['message']]
test['message'] = [i.strip().lower() for i in test['message']]
```

```python
train['message'] = [re.sub(r"([^\s])", r"\1 ", i) for i in train['message']]
test['message'] = [re.sub(r"([^\s])", r"\1 ", i) for i in test['message']]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

test_sentence = ["I am Harry", "I am Ye Bhone Lin!"]
vocab_size = 6
oov_token = "<OOV>"
tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_token)
tokenizer.fit_on_texts(test_sentence)
```

```python
word_index = tokenizer.word_index
print(word_index)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

sequences = tokenizer.texts_to_sequences(test_sentence)
print(sequences)
```

```python
max_length = 3
trunc_type = 'post'

padding = pad_sequences(sequences, maxlen=max_length, truncating=trunc_type,padding='post')
padding
```

```python
vocab_size = 500
max_length = 4182
trunc_type='post'        #put needed '0's for max length
oov_tok = "<OOV>"        #out of vocab


tokenizer = Tokenizer(num_words = vocab_size, oov_token = oov_tok)
tokenizer.fit_on_texts(train['message'])
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(train['message'])
padded = pad_sequences(sequences,maxlen=max_length, truncating=trunc_type,padding='post')
```

```python
import tensorflow as tf
embedding_dim = 64
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length = max_length),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(500, activation='relu'),
    tf.keras.layers.Dense(1000, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy', optimizer="adam", metrics=['accuracy'])
```

```python
epochs = 10
history = model.fit(padded, np.array(train['label']), epochs=epochs, verbose=2)
```

```python
testing_sequences = tokenizer.texts_to_sequences(test['message'])
testing_padded = pad_sequences(testing_sequences,maxlen=max_length, truncating=trunc_type,padding='post')
```

```python
y_predict = model.predict(testing_padded)
```

```python
y_predict
```

```python
y_pred = []
threshold = 0.5
for i in y_predict:
  current_val = i[0]
  if current_val >= threshold:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(test['label'],y_pred)
print('confusion_matrix')
print(confusion_matrix)
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
ans = y_pred
print('Accuracy Score',accuracy_score(test['label'],ans)*100,'%')
print('Precision Macro Score',precision_score(test['label'],ans,average = 'macro')*100,'%')
print('Recall_Score',recall_score(test['label'],ans, average = 'macro')*100,'%')
print('F1_Score',f1_score(test['label'],ans, average = 'macro')*100,'%')
```
